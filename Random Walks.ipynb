# Random Walks

This notebook will go through the implementation of a particular stochastic process known as a random walk, which comes from the concept of Markov Processes.

## Imports

```python
# Import Libraries
from typing import List
import numpy as np
import matplotlib.pyplot as plt
```

## Functions

```python
def simple_random_walk(p: float, num_iter: int, start_pos: 0 = int):
    # Store observations - note that the state space is the integers
    observations = [start_pos]
    for _ in range(num_iter):
        # Sample a uniform random variable
        proba = np.random.uniform(0, 1)
        # p defines increment up by 1
        if proba < p:
            start_pos += 1
        else:
            start_pos -= 1
        # Store observation
        observations.append(start_pos)
    # Return observations
    return observations
```

```python
def plot_random_walk(observations: List[List], title: str):
    # Plot multiple realisation paths for different experiments
    fig, ax = plt.subplots()
    for realisation in observations:
        ax.plot(realisation)
    ax.set_xlabel('Iteration')
    ax.set_ylabel('Path')
    ax.set_title(title)
```

## Results

```python
# Obtain multiple paths
observation_list = [simple_random_walk(p=0.25, num_iter=5000, start_pos=0) for _ in range(50)]
plot_random_walk(observations=observation_list, title= "Downward Biased Random Walk Trajectories")
```

```python
# Obtain multiple paths
observation_list = [simple_random_walk(p=0.5, num_iter=5000, start_pos=0) for _ in range(50)]
plot_random_walk(observations=observation_list, title= "Symmetric Random Walk Trajectories")
```

```python
# Obtain multiple paths
observation_list = [simple_random_walk(p=0.75, num_iter=5000, start_pos=0) for _ in range(50)]
plot_random_walk(observations=observation_list, title= "Upward Biased Random Walk Trajectories")
```

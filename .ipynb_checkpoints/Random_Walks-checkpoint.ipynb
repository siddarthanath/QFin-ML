# Random Walks

This notebook will go through the implementation of a particular stochastic process known as a random walk, which comes from the concept of Markov Processes.
\
For a theoretical understanding of Random Walks, please click the following link: ...

## Imports

```python
# Import Libraries
from typing import List
import numpy as np
import matplotlib.pyplot as plt
```

## Functions

```python
def simple_random_walk(prob: float, n: int, start_pos: 0 = int):
    # Pick random integer between -1 and 1 
    increments = np.random.choice([-1, 1], size=n, p=[1-prob, prob])
    # Return random walk
    rw = np.cumsum(increments)
    return np.insert(rw, 0, 0.0)
```

```python
def plot_random_walk(ax, observations: List[List], title: str):
    for realisation in observations:
        ax.plot(realisation)
    ax.set_xlabel('Iteration')
    ax.set_ylabel('Path')
    ax.set_title(title)
```

## Results

### Non-coin toss

```python
# Obtain multiple paths
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
observation_list = [simple_random_walk(prob=0.25, n=5000, start_pos=0) for _ in range(50)]
plot_random_walk(ax=ax, observations=observation_list, title= "Downward Biased Random Walk Trajectories")
```

Remember that $p$ is the probability that we increment up by $1$ - analogous to the coin-toss problem, it is $\mathbb{P}\left(H\right)$ and receiving £1. Additionally, $1-p$ is the probability that we increment down by $1$ - analogous to the coin-toss problem, it is $\mathbb{P}\left(T\right)$ and losing £1. Since $p=0.25$, it is more likely we obtain T and thus increment down/lose £1, hence almost all trajectories are downward pointing, with complete negative paths. Using this theory, it is clear to see why the other plots look the way they do.

```python
# Obtain multiple paths
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
observation_list = [simple_random_walk(prob=0.5, n=5000, start_pos=0) for _ in range(50)]
plot_random_walk(ax=ax, observations=observation_list, title= "Horizontal Unbiased Random Walk Trajectories")
```

```python
# Obtain multiple paths
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
observation_list = [simple_random_walk(prob=0.75, n=5000, start_pos=0) for _ in range(50)]
plot_random_walk(ax=ax, observations=observation_list, title= "Upward Biased Random Walk Trajectories")
```

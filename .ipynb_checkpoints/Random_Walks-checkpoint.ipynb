# Random Walks

This notebook will go through the implementation of a particular stochastic process known as a random walk, which comes from the concept of Markov Processes.
\

## Imports

```python
# Import Libraries
from typing import List
import numpy as np
import matplotlib.pyplot as plt
```

## Functions

```python
def simple_random_walk(prob: float, num_iter: int, start_pos: 0 = int, problem_type: "coin-toss" = str):
    # Store observations - note that the state space is the integers
    observations = [start_pos]
    for _ in range(num_iter):
        if problem_type == "coin-toss":
            # Pick random integer between -1 and 1 
            obs = np.random.choice([-1, 1], p=[1-prob, prob])
            start_pos += obs
        else:
            # Sample a uniform random variable
            random_proba = np.random.uniform(0, 1)
            # p defines increment up by 1
            start_pos = start_pos + 1 if random_proba < prob else start_pos - 1
        # Store observation
        observations.append(start_pos)
    # Return observations
    return observations
```

```python
def plot_random_walk(ax, observations: List[List], title: str):
    for realisation in observations:
        ax.plot(realisation)
    ax.set_xlabel('Iteration')
    ax.set_ylabel('Path')
    ax.set_title(title)
```

## Results

### Non-coin toss

```python
# Obtain multiple paths
fig, axs = plt.subplots(1, 2, figsize=(15, 7.5))
observation_list = [simple_random_walk(prob=0.25, num_iter=5000, start_pos=0, problem_type="N/A") for _ in range(50)]
plot_random_walk(ax=axs[0], observations=observation_list, title= "Downward Biased Random Walk Trajectories (Non-coin toss)")
observation_list = [simple_random_walk(prob=0.25, num_iter=5000, start_pos=0, problem_type="coin-toss") for _ in range(50)]
plot_random_walk(ax=axs[1], observations=observation_list, title= "Downward Biased Random Walk Trajectories (Coin toss)")
```

Remember that $p$ is the probability that we increment up by $1$ - analogous to the coin-toss problem, it is $\mathbb{P}\left(H\right)$ and receiving £1. Additionally, $1-p$ is the probability that we increment down by $1$ - analogous to the coin-toss problem, it is $\mathbb{P}\left(T\right)$ and losing £1. Since $p=0.25$, it is more likely we obtain T and thus increment down/lose £1, hence almost all trajectories are downward pointing, with complete negative paths. Using this theory, it is clear to see why the other plots look the way they do.

```python
# Obtain multiple paths
fig, axs = plt.subplots(1, 2, figsize=(15, 7.5))
observation_list = [simple_random_walk(prob=0.5, num_iter=5000, start_pos=0, problem_type="N/A") for _ in range(50)]
plot_random_walk(ax=axs[0], observations=observation_list, title= "Downward Biased Random Walk Trajectories (Non-coin toss)")
observation_list = [simple_random_walk(prob=0.5, num_iter=5000, start_pos=0, problem_type="coin-toss") for _ in range(50)]
plot_random_walk(ax=axs[1], observations=observation_list, title= "Downward Biased Random Walk Trajectories (Coin toss)")
```
